```python
import pandas as pd
```

```python
df=pd.read_csv('Dataset/train.csv')
```

```python
df
```

```python
df.drop(['Loan_ID','Gender','Loan_Amount_Term'],axis=1, inplace=True)
```

```python
df.isnull().any()
```

```python
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df
```

```python
df = pd.get_dummies(df,columns=['Education'], drop_first=True)
df = pd.get_dummies(df,columns=['Married'], drop_first=True)
df = pd.get_dummies(df,columns=['Self_Employed'], drop_first=True)
df = pd.get_dummies(df,columns=['Property_Area'], drop_first=True)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df['Loan_Status']=le.fit_transform(df['Loan_Status'])
```

```python
def normalizing(text):
    if text=='3+':
        return '3'
    else:
        return text
```

```python
df['Dependents']=df['Dependents'].apply(normalizing)
```

```python
df
```

```python
X=df.iloc[:,[0,1,2,3,4,6,7,8,9,10]]
y=df.iloc[:,[5]]
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(random_state=42)
```

```python
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc.fit(X, y)
```

```python
CV_rfc.best_params_
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(criterion='gini',max_depth=4,max_features='auto',n_estimators=200)
model.fit(X,y)
```

```python
model.predict([[0,5849,0.0,146.412162,1.0,0,0,0,0,1]])
```

```python
import pickle

file=open('Model/Prediction.pkl','wb')
pickle.dump(model,file)
file.close()
```
