```python
import pandas as pd
```

```python
df=pd.read_csv('Dataset/test.csv')
```

```python
df
```

```python
df.drop(['Loan_ID','Gender','Loan_Amount_Term'],axis=1, inplace=True)
```

```python
df.isnull().any()
```

```python
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Credit_History']=df['Credit_History'].fillna(df['Credit_History'].mode()[0])
```

```python
df
```

```python
df = pd.get_dummies(df,columns=['Education'], drop_first=True)
df = pd.get_dummies(df,columns=['Married'], drop_first=True)
df = pd.get_dummies(df,columns=['Self_Employed'], drop_first=True)
df = pd.get_dummies(df,columns=['Property_Area'], drop_first=True)
```

```python
def normalizing(text):
    if text=='3+':
        return '3'
    else:
        return text
```

```python
df['Dependents']=df['Dependents'].apply(normalizing)
```

```python
df
```

```python
import pickle

file=open('Model/Prediction.pkl','rb')
model=pickle.load(file)
file.close()
```

```python
for i in range(0,367):
    dep= int(df['Dependents'][i])
    appInc=int(df['ApplicantIncome'][i])
    CoApp=float(df['CoapplicantIncome'][i])
    Loan=float(df['LoanAmount'][i])
    cred=float(df['Credit_History'][i])
    edu=int(df['Education_Not Graduate'][i])
    married=int(df['Married_Yes'][i])
    self=int(df['Self_Employed_Yes'][i])
    prop=int(df['Property_Area_Urban'][i])
    urb=int(df['Property_Area_Semiurban'][i])

    res=model.predict([[dep,appInc,CoApp,Loan,cred,edu,married,self,prop,urb]])
    if res[0]==0:
        res='N'
    else:
        res='y'
        
    df['Status']=res
    
    
```
